```python
# Linear Regression with IRIS Dataset

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

iris_data = pd.read_csv("IRIS.csv") #Loading IRIS dataset

X = iris_data.iloc[:,0:4] #First four columns of data
Y = iris_data.iloc[:,4] #Target variable i.e. species of flowers(last column of data)

y = np.array(Y) #Converting pandas object Y to a numpy array 

# Plot species with features petal length and width
sns.lmplot(x="petal_width", y="petal_length", data=iris_data, hue="species", fit_reg=False, legend=False)

plt.legend()
plt.show()

# Plot species with features sepal length and width
sns.lmplot(x="sepal_width", y="sepal_length", data=iris_data, hue="species", fit_reg=False, legend=False)

plt.legend()
plt.show()

# Encoding three species of flower: "Iris-setosa" as 1, "Iris-versicolor" as 2 and "Iris-virginica" as 3
for i in range(len(y)):
    if(y[i]=='Iris-setosa'):
        y[i]=1
    elif(Y[i]=='Iris-versicolor'):
        y[i]=2
    else:
        y[i]=3      
        
# Splitting data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=13) 

lr1 = linear_model.LinearRegression()
lr1.fit(X_train, y_train) # Fitting data on linear regression

y_predict = lr1.predict(X_test) # Predicting results on test data

print("coefficient:", lr1.coef_) # Coefficients
print("MSE: ",mean_squared_error(y_test, y_predict)) # Mean Squared Error
print("R2_score:", r2_score(y_test, y_predict)) # R2 score
```
